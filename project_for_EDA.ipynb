```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=data,x='Year',y='Price',ci=None,marker='o',estimator=np.mean,label='Mean Salary')
plt.title('Mean Price Over Years', fontsize=14, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel(' price', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
top_models = data.groupby("Model")["Price"].mean().nlargest(10)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(
    data=data[data["Model"].isin(top_models.index)],
    x="Model",
    y="Price",
    hue='Make',
    estimator=np.mean,
    order=top_models.index,
    palette="muted")

plt.xlabel("Car Model", fontsize=12)
plt.ylabel(" Avarage Price", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```

```python
top_body_type = data.groupby("Body Type")["Price"].mean().nlargest(10)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(
    data=data[data["Body Type"].isin(top_body_type.index)],
    x="Body Type",
    y="Price",
    estimator=np.mean,
    order=top_body_type.index,
    palette="viridis")

plt.title("Top 10 Car Body Type by  Price", fontsize=14, fontweight="bold")
plt.xlabel("Body Type", fontsize=12)
plt.ylabel(" Price", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```

```python
top_makes = data["Make"].value_counts().nlargest(10)
```

```python

plt.figure(figsize=(12, 6))
sns.countplot(
    data=data[data["Make"].isin(top_makes.index)],
    x="Make",
    order=top_makes.index,
    palette="viridis"
)
plt.title("Top 10 Car Makes by Count", fontsize=14, fontweight="bold")
plt.xlabel("Car Make", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```
